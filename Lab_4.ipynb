6. С помощью модуля numPy реализуйте следующие операции: 

* умножение произвольных матриц А (размерности 3х5) и В (5х2); 
* умножение матрицы (5х3) на трехмерный вектор; 
* решение произвольной системы линейных уравнений; 
* расчет определителя матрицы; 
* получение обратной и транспонированной матриц.

Также продемонстрируйте на примере матрицы 5х5 тот факт, что
определитель равен произведению собственных значений матрицы.

```python
import numpy

def mult_matrix():
    a = numpy.arange(3 * 5).reshape((3, 5))
    b = numpy.arange(5 * 2).reshape((5, 2))

    print(f'A:\n{a}')
    print(f'B:\n{b}')
    print(f'A*B:\n {a @ b} \n')

def mult_matrix_vector():
    matrix = numpy.arange(5 * 3).reshape((5, 3))
    vector = numpy.array([1, 0.5, -1], dtype=float)
    print('M*V:\n',matrix @ vector,'\n')

def solve_sle():
    matrix = numpy.array([[6., 7.], [3., -5.]])
    vector = numpy.array([17., 6.])
    print('6x + 7y = 17')
    print('3x - 5y = 6')
    print('       x        y')
    print(f'{numpy.linalg.solve(matrix, vector)}\n')

def determinant_matrix():
    matrix = numpy.random.randint(25, size = (5, 5))
    print(f'M:\n{matrix}')
    print(f' det = {numpy.linalg.det(matrix)} \n')

def inv_trans_matrix():
    matrix = numpy.array([[0, 5, 8], [6, 5, 2], [9, 3, 15]])
    matrix_inverse = numpy.linalg.inv(matrix)
    matrix_transposed = matrix.transpose()
    print(f'M:\n{matrix}')
    print(f'Invented M:\n{matrix_inverse}')
    print(f'Transposed M:\n{matrix_transposed}')

mult_matrix()
mult_matrix_vector()
solve_sle()
determinant_matrix()
inv_trans_matrix()

```

7. Выберите произвольную дифференцируемую и интегрируемую
функцию одной переменной. С помощью модуля symPy найдите и
отобразите ее производную и интеграл в аналитическом и
графическом виде. Напишите код для решения произвольного
нелинейного уравнения и системы нелинейных уравнений.

```python
import sympy

def diff_and_integ():
    x = sympy.Symbol('x')
    func = 7*x**4 

    display_function(func, 'Функция')
    display_function(sympy.diff(func), 'Производная')
    display_function(sympy.integrate(func), 'Интеграл')

def display_function(func, name):
    print(f'\n{name}:\n')
    sympy.pprint(func)
    sympy.plot(func)

def solve_sne(equations):
    if len(equations) == 1:
        return sympy.solve(equations[0])
    else:
        return sympy.solve_poly_system(equations)


def sne():
    x = sympy.Symbol('x')
    y = sympy.Symbol('y')

    a = sympy.Equality(-5, - x * y - 3*y)
    sympy.pprint(a)
    b = sympy.Equality(6, x * y + x**2)
    sympy.pprint(b)
    sympy.pprint(solve_sne([a, b]))
    
    c = sympy.Equality(x ** 3, 8)
    sympy.pprint(c)
    sympy.pprint(solve_sne([c]))


diff_and_integ()
print()
sne()
```

8. С помощью модуля pandas отобразите:

    1) 10 самых маленьких и самых больших стран мира по территории;

    2) 10 самых маленьких и самых больших стран мира по населению;

    3) все франкоязычные страны мира;

    4) только островные государства;

    5) все страны, находящиеся в южном полушарии.

Сгруппируйте страны:
    по первой букве;
    по населению;
    по территории.
Программно сохраните в таблицу Excel.
все страны с выборочной информацией:
название, столица, население, территория, валюта, широта, долгота.

```python
import csv
import pandas

def export_to_excel(table):
    names = pandas.Series([c.split(',')[0] for c in table.name])
    names.name = 'name'
    lat, lng = zip(*[l.split(',')
                     if isinstance(l, str)
                     else ['nan', 'nan']
                     for l in table.latlng])
    lat, lng = map(pandas.Series, (lat, lng))
    lat.name = 'latitude'
    lng.name = 'longitude'
    for_export = pandas.concat([names, table[['capital', 'ccn3', 'area', 'currencies']], lat, lng], axis=1)
    with pandas.ExcelWriter('export.xls') as excel_writer:
        for_export.to_excel(excel_writer)


table = pandas.read_csv('countries.csv', ',')

print('↓ Area'.center(70, '-'))
print(table.nlargest(10, columns='area')[['area', 'name']])  

print('↑ Area'.center(70, '-'))
print(table.nsmallest(10, ['area'])[['area', 'name']])

print('↓ Population density'.center(60, '-'))
print(table.nlargest(10, ['ccn3'])[['ccn3', 'name']])  

print('↑ Population density'.center(60, '-'))
print(table.nsmallest(10, ['ccn3'])[['ccn3', 'name']])

print('French countries'.center(80, '-'))
print(table[table.languages == 'French'][['languages', 'area', 'name']])  

print('Islands'.center(60, '-'))
print(table[table.borders.isnull()][['name']])  
export_to_excel(table)

print('Southern Hemisphere'.center(70, '-'))
print(table.where(pandas.Series([float(str(l).split(',')[0]) < 0 for l in table.latlng])).name.dropna())

for i, group in table.groupby(table.area):
    print(f'{str(i)}: ')
    for name in sorted(group.name):
        print('·' + name.split(',')[0])

for i, group in table.groupby([d[0] for d in table.name]):  
    print(f'{str(i)}: ')
    for name in sorted(group.name):
        print('·' + name.split(',')[0])
```
